```python
import numpy as np

#Assignment 1
##Q1 Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.
arr = np.random.randint(1, 21, size=(5, 5))
arr[:, 2] = 1
print("Q1\n",arr)

##Q2 Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.
arr2 = np.arange(1, 17).reshape(4, 4)
np.fill_diagonal(arr2, 0)
print("\nQ2\n",arr2)
```

```python
## Assignment 2: Array Indexing and Slicing

#1. Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
arr3 = np.arange(1, 37).reshape(6, 6)
sub_arr = arr3[2:5, 1:4]
print("Q1\n", sub_arr)

#2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.
arr4 = np.random.randint(1, 21, size=(5, 5))
border_elements = np.concatenate([
    arr4[0, :],           # top row
    arr4[-1, :],          # bottom row
    arr4[1:-1, 0],        # left column (excluding corners)
    arr4[1:-1, -1]        # right column (excluding corners)
])
print("Q2\n", border_elements)
```

```python
## Assignment 3: Array Operations

#1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
arr_a = np.random.randint(1, 21, size=(3, 4))
arr_b = np.random.randint(1, 21, size=(3, 4))

add = arr_a + arr_b
sub = arr_a - arr_b
mul = arr_a * arr_b
div = arr_a / arr_b

print("Q1  Array A:\n", arr_a)
print("Array B:\n", arr_b)
print("Addition:\n", add)
print("Subtraction:\n", sub)
print("Multiplication:\n", mul)
print("Division:\n", div)

#2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.
arr5 = np.arange(1, 17).reshape(4, 4)
row_sum = arr5.sum(axis=1)
col_sum = arr5.sum(axis=0)
print("\nQ2  Row-wise sum:\n", row_sum)
print("Column-wise sum:\n", col_sum)
```

```python
### Assignment 4: Statistical Operations

#1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
arr6 = np.random.randint(1, 21, size=(5, 5))
mean = np.mean(arr6)
median = np.median(arr6)
std = np.std(arr6)
var = np.var(arr6)
print("Q1\nArray:\n", arr6)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
print("Variance:", var)

#2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).
arr7 = np.arange(1, 10).reshape(3, 3)
mean7 = np.mean(arr7)
std7 = np.std(arr7)
normalized_arr7 = (arr7 - mean7) / std7
print("\nQ2\nOriginal Array:\n", arr7)
print("Normalized Array:\n", normalized_arr7)
```

```python
## Assignment 5: Broadcasting

#1. Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.
arr8 = np.random.randint(1, 21, size=(3, 3))
row_add = np.random.randint(1, 21, size=(3,))
result1 = arr8 + row_add
print("Q1\nArray:\n", arr8)
print("1D Array to Add:\n", row_add)
print("Result:\n", result1)

#2. Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.
arr9 = np.random.randint(1, 21, size=(4, 4))
col_sub = np.random.randint(1, 21, size=(4,))
result2 = arr9 - col_sub
print("Q2\nArray:\n", arr9)
print("1D Array to Subtract:\n", col_sub)
print("Result:\n", result2)
```
